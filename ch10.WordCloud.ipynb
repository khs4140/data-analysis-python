```python
from wordcloud import WordCloud
from collections import Counter
from konlpy.tag import Hannanum
```

    - open()
        1. read() : 텍스트 파일 전체를 하나의 문자열로 읽음
        2. readline() : 텍스트 파일의 한 줄만 문자열로 읽음
        3. readlines() : 텍스트 파일 전체를 한 줄씩 문자열로 읽어와 리스트 형태로 변환

```python
data = open('./dataset/ch8.txt', encoding='UTF-8').read()

data
```

```python
n = Hannanum().nouns(data)
n
```

```python
## 단어 변경

print(n.index('재구'))
n[14] = '재구매'
```

```python
data_2 = [item for item in n if len(item) >= 2] # 2글자 이상만 추출

cnt = Counter(data_2)
cnt.most_common()[:10] # 빈도수가 많은 순서대로 표시
```

```python
wc = WordCloud(font_path='AppleGothic',background_color='black')
```

```python
cloud = wc.generate_from_frequencies(cnt)

import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'AppleGothic'

plt.figure(figsize=(20, 12))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```
