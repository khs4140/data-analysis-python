```python
import pandas as pd

from sklearn.datasets import load_iris

iris = load_iris()

print(f'iris.data : {iris.data}')
print(f'iris.feature : {iris.feature_names}')
print('--')
print(f'iris.target : {iris.target}')
print(f'iris.target_names : {iris.target_names}')
```

```python
dir(iris)
```

```python
type(iris.data)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names,)

df['species'] = iris.target
df.head()
```
